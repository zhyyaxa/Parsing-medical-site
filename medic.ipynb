```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
name = list()
spec = list()
years = list()
list_div.clear()
```

```python
for number_page in range(1,3):
    link = 'https://prodoctorov.ru/biysk/pediatr/?page=2'#+str(number_page)
    page_text = requests.get(link).text
    soup = BeautifulSoup(page_text, 'lxml')
    list_div = soup.find_all('div',attrs = {'class' : ['b-doctor-card']})
    print(len(list_div))
     # 'b-doctor-card__name-surname',b-doctor-card__spec,b-doctor-card__experience-years
    #print(list_div)
```

```python
    for doctor in list_div:
        name.append(doctor.find_all('span', class_ = 'b-doctor-card__name-surname')[0].get_text(strip = True))
        spec.append(doctor.find_all('div', class_ = 'b-doctor-card__spec')[0].get_text(strip = True))
        years.append(doctor.find_all('div', class_ = 'b-doctor-card__experience')[0].get_text(strip = True))
```

```python
df1 =          ({'Врач' : name,
               'Специальность' : spec,
               'Стаж' : years
              })
df2 = pd.DataFrame.from_dict(df1, orient = 'index')
df2 = df2.transpose()
df2.loc[1:3]
df2 = df2.style.set_properties(**{
    'background-color': 'white',
    'font-size': '10pt'})
```
